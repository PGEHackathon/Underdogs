```python
index_day_df = day_df.set_index(["Well_ID", "AL_Key"])
```

```python
days_to_failure = day_df.groupby(["Well_ID", "AL_Key"])["OIL"].count()
days_to_failure
```

```python
L30_daily_df = []
for _iter in tqdm(index_day_df.index.unique()):
    sub_df = index_day_df.loc[_iter]
    L30_daily_df.append(sub_df.tail(30))
    
L30_daily_df = pd.concat(L30_daily_df)
```

```python
B30_daily_df = []
for _iter in tqdm(index_day_df.index.unique()):
    sub_df = index_day_df.loc[_iter]
    B30_daily_df.append(sub_df.head(30))
    
B30_daily_df = pd.concat(B30_daily_df)
```

```python
L30_daily_df = L30_daily_df.drop(['Lower_Limit', 'Power_Difference', 'ESP Data - Motor Winding Temperature', 'Drive_Power','ESP Data - Output Frequency',
'Gas_through_Annulus_Intake',
'ESP Data - Drive Voltage',
'Discharge_Pressure',
'Gas_Saturation_at_Discharge',
'Pump_Average_Pressure',
'Pump_Power',
'Power_Ratio'], axis = 1)
```
